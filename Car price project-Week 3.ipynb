## Data preparation

```python
df1.columns = df1.columns.str.lower().str.replace(' ', '_')

categorical_columns = list(df1.dtypes[df1.dtypes == 'object'].index)

for c in categorical_columns:
    df1[c] = df1[c].str.lower().str.replace(' ', '_')
```

```python
df1.isna().sum()
```

```python
df1.engine_hp = df1.engine_hp.fillna(0)
```

```python
df1.engine_cylinders = df1.engine_cylinders.fillna(0)
```

```python
df1.isna().sum()
```

```python
df1.rename(columns={'msrp': 'price'}, inplace=True)
```

### Making price binary

### Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
df1_full_train, df1_test = train_test_split(df1, test_size=0.2, random_state=42)
```

```python
len(df1_full_train), len(df1_test)
```

```python
df1_train, df1_val = train_test_split(df1_full_train, test_size=0.25, random_state=42)
```

```python
len(df1_train), len(df1_val), len(df1_test)
```

```python
df1_train = df1_train.reset_index(drop=True)
df1_val = df1_val.reset_index(drop=True)
df1_test = df1_test.reset_index(drop=True)
```

```python
y_train = df1_train.above_average.values
y_val = df1_val.above_average.values
y_test = df1_test.above_average.values
```

## Question 4

### One-hot encoding

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)

train_dicts = df1_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df1_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
model.intercept_[0]
```

```python
model.coef_[0].round(3)
```

```python
y_pred = model.predict_proba(X_val)[:,1]
```

```python
above_average_decision = (y_pred >= 0.5)
```

```python
round((y_val == above_average_decision).mean(),2)
```
